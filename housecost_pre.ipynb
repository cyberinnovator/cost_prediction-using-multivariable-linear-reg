```python
df=pd.read_csv('Bangalore  house data.csv')
df
```

```python
df.columns
```

```python
df['balcony'].fillna(0, inplace=True)  # Fill missing balconies with 0
df['bath'].fillna(df['bath'].median(), inplace=True)  # Fill missing baths with median
df['total_sqft'] = pd.to_numeric(df['total_sqft'], errors='coerce') # Convert 'total_sqft' to numeric, replacing non-numeric with NaN
df['total_sqft'].fillna(df['total_sqft'].median(), inplace=True)  # Fill missing sqft using the median of the numeric values
df.dropna(subset=['size', 'area_type'], inplace=True)  # Drop rows with missing categorical data
```

```python
df['size'] = df['size'].str.extract('(\d+)').astype(float)  # Extract numeric part
```

```python
def convert_sqft(sqft):
    if '-' in str(sqft):
        vals = sqft.split('-')
        return (float(vals[0]) + float(vals[1])) / 2
    try:
        return float(sqft)
    except ValueError:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_sqft)
```

```python
df.dropna(subset=['total_sqft'], inplace=True)
```

```python
le = LabelEncoder()
df['area_type'] = le.fit_transform(df['area_type'])
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['total_sqft','bath','size','balcony','area_type']],df.price)
```
